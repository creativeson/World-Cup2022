# Create Continent Table

```python
import requests
import pandas as pd

url = 'https://statisticstimes.com/geography/countries-by-continents.php'

header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}
r = requests.get(url, headers=header)
dfs = pd.read_html(r.text)
df_continent_raw = dfs[2]
```

```python
#select columns we want
df_continent_raw = df_continent_raw[['Country or Area', 'ISO-alpha3 Code', 'Region 1', 'Continent']]
```

```python
#df_continent_raw[df_continent_raw['ISO-alpha3 Code'] == 'DEU']
df_continent_raw
```

### unify the country code

```python
df_continent_raw.loc[df_continent_raw['ISO-alpha3 Code'] == 'NLD','ISO-alpha3 Code']='NED'
df_continent_raw.loc[df_continent_raw['ISO-alpha3 Code'] == 'GBR','ISO-alpha3 Code']='ENG'
df_continent_raw.loc[df_continent_raw['ISO-alpha3 Code'] == 'CRI','ISO-alpha3 Code']='CRC'
df_continent_raw.loc[df_continent_raw['ISO-alpha3 Code'] == 'DEU','ISO-alpha3 Code']='GER'
df_continent_raw.loc[df_continent_raw['ISO-alpha3 Code'] == 'HRV','ISO-alpha3 Code']='CRO'
df_continent_raw.loc[df_continent_raw['ISO-alpha3 Code'] == 'CHE','ISO-alpha3 Code']='SUI'
```

```python
df_continent_raw[df_continent_raw['ISO-alpha3 Code'] == 'ENG']
```

```python
df_continent_raw
```

### add wales data as new row

```python
df_continent_raw.loc[len(df_continent_raw.index)] = ['Wales', 'WAL', 'Northern Europe', 'Europe']
```

```python
df_continent_raw.rename(columns = {'ISO-alpha3 Code':'Code'}, inplace = True)
```

```python
df_continent_raw
```

```python
#only select countries that are in world cup
name= ['QAT','ECU','SEN','NED',
       'ENG','IRN','USA','WAL',
       'ARG','SAU','MEX','POL',
       'FRA','AUS','DNK','TUN',
       'ESP','COI','GER','JPN',
       'BEL','CAN','MAR','CRO',
       'BRA','SRB','SUI','CMR',
       'PRT','GHA','URY','KOR']

code = df_continent_raw['Code'].isin(name)
filtered_df = df_continent_raw[code]
```

```python
filtered_df['Count'] = 1
filtered_df.head()
```

```python
#Austrulia belong to AFC now, so we have to modify the data
wc_by_continent = pd.DataFrame(filtered_df.groupby('Continent', as_index = False)['Count'].sum())
wc_by_continent1 = wc_by_continent.drop(axis = 'index', index = 4)
wc_by_continent1['Count'][1] = wc_by_continent1['Count'][1] + 1 
```

```python
wc_by_continent1
```

```python
#groupby continent
ax = sns.barplot(x = 'Continent', y = 'Count', data = wc_by_continent1.sort_values(by = 'Count'), palette ='Set2')
ax.set_xlabel("Nation", fontsize = 20)
ax.set_ylabel("Count", fontsize = 20)
plt.rcParams['figure.figsize'] = (10, 3)
plt.title('World Cup Nations by continent', fontsize = 20)
```

```python
#groupby region1
wc_by_continent = pd.DataFrame(filtered_df.groupby('Region 1', as_index = False)['Count'].sum())
ax = sns.barplot(x = 'Region 1', y = 'Count', data = wc_by_continent.sort_values(by = 'Count'), palette ='Set2')
ax.set_xlabel("Nation", fontsize = 20)
ax.set_ylabel("Count", fontsize = 20)
plt.rcParams['figure.figsize'] = (15, 5)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
plt.title('World Cup Nations by Region', fontsize = 20)
```

```python
#only select the players whose nations are in the world cup 
df_two = pd.merge(filtered_df, df, left_on="Code", right_on="Nation")
```

```python
#Select the stat we want observe
wc_df = df_two[['Country or Area', 'Nation', 'Region 1', 
        'Continent', 'Player', 'Pos', 'Squad', 
        'Comp', 'Age', 'Min', 'Goals', 'Shots','ShoPK',
        'Assists', 'PasTotDist', 'CrsPA', 'PasCrs',
        'Off', 'ScaDrib', 'GCA', 'TklWon', 'PresSucc',
        'Touches', 'Err', 'Int', 'Fld', 'AerWon','CrdY'
       ]]
```

```python
wc_df
```

```python
#wc_df.isna().sum()
wc_df['Pos'].unique()
```

```python
pos = wc_df.groupby('Pos').size().sort_values(ascending = False)
pos.plot(kind = 'bar', figsize = (12, 6), color = sns.color_palette('magma'))
```

```python
#create a bar chart base on their nation
nation = wc_df.groupby('Nation').size().sort_values(ascending = False)
nation.plot(kind = 'bar', figsize = (15, 8), color = sns.color_palette('magma'))
```

```python
# player base on age group 
under20 = wc_df[wc_df['Age'] <= 20]
age20_25 = wc_df[(wc_df['Age'] > 20) &(wc_df['Age'] <= 25)]
age25_30 = wc_df[(wc_df['Age'] > 25) &(wc_df['Age'] <= 30)]
age30_35 = wc_df[(wc_df['Age'] > 30) &(wc_df['Age'] <= 35)]
above35 = wc_df[wc_df['Age'] > 35]
```

```python
x = np.array([under20['Player'].count(), age20_25['Player'].count(),
            age25_30['Player'].count(), age30_35['Player'].count(), 
            above35['Player'].count()])
label = ['<=20', '20-25', '25-30', '30-35', '>35']
plt.title('Players With Age Group', fontsize = 20)
plt.pie(x, labels = label, autopct = "%.1f%%")
plt.show()
```

```python
#average age of players in each country
plt.figure(figsize = (16, 7))
sns.boxplot(x = 'Nation', y = 'Age', data = wc_df)
```

```python
#Argentina players average age is the oldest, and USA players average age is the youngest
num_player = wc_df.groupby('Nation').size()
data = (wc_df.groupby('Nation')['Age'].sum() / num_player)
data.sort_values(ascending = False)
```

```python
# create new columns
wc_df['ATK']= wc_df['Goals'] + wc_df['Assists'] # contribution goal and assists each 90 mins
wc_df['Go'] = round(wc_df['Goals']*wc_df['Min'] /90, 0).astype(int)
wc_df['Ast'] = round(wc_df['Assists']*wc_df['Min'] /90, 0).astype(int)  #改變成整數型態
wc_df['PK'] = round(wc_df['ShoPK']*wc_df['Min'] /90, 0).astype(int)
wc_df['YellowCard'] = round(wc_df['CrdY']*wc_df['Min'] /90, 0).astype(int)
```

```python
#top 20 ast
top_20_ast = wc_df[['Player', 'Nation', 'Squad', 'Min', 'Ast']].nlargest(n = 20, columns = 'Ast')
top_20_ast
```

```python
#top 20 Goal
top_20_go = wc_df[['Player', 'Nation', 'Squad', 'Min', 'Go']].nlargest(n = 20, columns = 'Go')
top_20_go
```

```python
#top 20 player has highest ATK rate each 90
wc1000_df = wc_df[wc_df['Min'] > 500] #set a min bar to exclude outliner
top_20_go = wc1000_df[['Player', 'Nation', 'Squad','Age', 'Min', 'ATK']].nlargest(n = 20, columns = 'ATK')
top_20_go
```

```python
#Top 10 players with  most yellow cards
wc_yellow = wc_df.sort_values(by = 'YellowCard', ascending = False)[:20]
plt.figure(figsize = (20, 6))
plt.title('Players with the most yellow cards')
c = sns.barplot(x = wc_yellow['Player'], y = wc_yellow['YellowCard'], label = 'Players', color = 'yellow')
plt.ylabel('Number of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(), rotation = 45)
c
```
